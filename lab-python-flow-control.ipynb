```python
products = ["t-shirt", "mug", "hat", "book", "keychain"]
inventory = {}

print("Please, enter the avaliable quantity of each product")

for i in products:
    inp = int(input("{}: ".format(i)))
    inventory[i] = inp
print("The following products are availiable in the stock:", inventory)

customer_orders = set() # Func. set() helps to identify Set instead of Dictionary -> Dictionary has no .add() method


while True:
    question = input("Do you want to add product name to the order? If 'yes', please, enter the product name from the above list. If 'no', please, enter 'no'").lower()
    if question == "no":
        break
    if question in products:
        if inventory[question] > 0:
            customer_orders.add(question) 
        else:
            print("Sorry, this product is out of stock")
    else:
        print("Please, enter the name from the list", products)

print("Your order contains this products:", customer_orders)
```

```python
tpo = len(customer_orders)
ppo = int(len(customer_orders)*100/len(products))

order_status = (tpo, ppo)
 
print("Total Products Ordered: {}".format(order_status[0]))
print("Percentage of Products Ordered: {}%".format(order_status[1]))
```
